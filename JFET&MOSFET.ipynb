```python
#Question 2
import numpy as np
import matplotlib.pyplot as plt

import shapely
from shapely.geometry import LineString, Point


Idss=10
Vp=-4
n=100
h=abs(Vp)/n
Rs=0.75
Vgs=np.arange(Vp,0+h,h)

def Id(Vgs):
    return Idss*(1-Vgs/Vp)**2
def ID(Vgs):
    return -Vgs/Rs

f = plt.figure()
ax = f.add_subplot(111)
ax.yaxis.tick_right()
ax.yaxis.set_label_position("right")

plt.grid()
plt.axvline(x=0,c='black')
plt.axhline(y=0,c='black')
plt.plot(Vgs,Id(Vgs), c='blue')
plt.plot(Vgs,ID(Vgs), c='black')

plt.title('Graphical Analysis ($I_{D}$ Against $V_{GS}$)')
plt.xlabel('$V_{GS}$ (V)')
ax.set_ylabel('$I_{D}$ (mA)')

line1 = LineString(np.column_stack((Vgs, ID(Vgs))))
line2 = LineString(np.column_stack((Vgs, Id(Vgs))))
intersection = line2.intersection(line1)


plt.plot(*intersection.xy, 'ro')
plt.show()    
x, y = intersection.xy
print(x,y)

    
    
```

```python
#Question 6
import numpy as np
import matplotlib.pyplot as plt

import shapely
from shapely.geometry import LineString, Point


n=100
h=1/n
Rd=1.2
Rs=0.51
Vdd=22

Idson=5
Vgson=7
Vgsth=4

Vgs=np.arange(Vgsth,10+h,h)
VGS=np.arange(0,Vdd+h,h)

k=Idson/(Vgson-Vgsth)**2

def Id(Vgs):
    return k*(Vgs-Vgsth)**2
def ID(VGS):
    return -(VGS-Vdd)/(Rs+Rd)

f = plt.figure()
ax = f.add_subplot(111)
ax.yaxis.tick_right()
ax.yaxis.set_label_position("right")

plt.grid()
plt.axvline(x=0,c='black')
plt.axhline(y=0,c='black')
plt.plot(Vgs,Id(Vgs), c='blue')
plt.plot(VGS,ID(VGS), c='black')

plt.title('Graphical Analysis ($I_{D}$ Against $V_{GS}$)')
plt.xlabel('$V_{GS}$ (V)')
ax.set_ylabel('$I_{D}$ (mA)')

line1 = LineString(np.column_stack((VGS, ID(VGS))))
line2 = LineString(np.column_stack((Vgs, Id(Vgs))))
intersection = line2.intersection(line1)


plt.plot(*intersection.xy, 'ro')
plt.show()    
x, y = intersection.xy
print(x,y)


print(k)
    
```
